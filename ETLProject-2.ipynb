### Store 2nd CSV data into a DataFrame

```python
#Load in Census data

census = pd.read_csv("acs2015_census_tract_data.csv", index_col=0)
census.head(10)
```

```python
census_data = census[["State", "TotalPop","Men","Women","Hispanic","White","Black","Native",
          "Asian","Pacific","Income","IncomePerCap","Poverty",
         "Professional","Service","Office","Construction","Production",
         "Employed","PrivateWork","PublicWork","SelfEmployed","FamilyWork","Unemployment"]]
census_data.head()
```

```python
census_data.groupby("State")
census_data.head()
```

```python
population = census_data.groupby(["State"]).sum()["TotalPop"]
men = census_data.groupby(["State"]).sum()["Men"]
women = census_data.groupby(["State"]).sum()["Women"]
hispanic = census_data.groupby(["State"]).mean()["Hispanic"]
white = census_data.groupby(["State"]).mean()["White"]
black = census_data.groupby(["State"]).mean()["Black"]
native = census_data.groupby(["State"]).mean()["Native"]
asian = census_data.groupby(["State"]).mean()["Asian"]
pacific = census_data.groupby(["State"]).mean()["Pacific"]
income = census_data.groupby(["State"]).mean()["Income"]
incomepercap = census_data.groupby(["State"]).mean()["IncomePerCap"]
poverty = census_data.groupby(["State"]).mean()["Poverty"]
professional = census_data.groupby(["State"]).mean()["Professional"]
service = census_data.groupby(["State"]).mean()["Service"]
office = census_data.groupby(["State"]).mean()["Office"]
construction = census_data.groupby(["State"]).mean()["Construction"]
production = census_data.groupby(["State"]).mean()["Production"]
employed = census_data.groupby(["State"]).mean()["Employed"]
privatework = census_data.groupby(["State"]).mean()["PrivateWork"]
publicwork = census_data.groupby(["State"]).mean()["PublicWork"]
selfemployed = census_data.groupby(["State"]).mean()["SelfEmployed"]
familywork = census_data.groupby(["State"]).mean()["FamilyWork"]
unemployment = census_data.groupby(["State"]).mean()["Unemployment"]

census_summary = pd.DataFrame({"TotalPop":population,
                              "Men":men,
                              "Women":women,
                              "Hispanic":hispanic,
                              "White":white,
                              "Black":black,
                              "Native":native,
                              "Asian":asian,
                              "Pacific":pacific,
                              "Income":income,
                              "IncomePerCap":incomepercap,
                              "Poverty":poverty,
                              "Professional":professional,
                              "Service":service,
                              "Office":office,
                              "Construction":construction,
                              "Production":production,
                              "Employed":employed,
                              "PrivateWork":privatework,
                              "PublicWork":publicwork,
                              "SelfEmployed":selfemployed,
                              "FamilyWork":familywork,
                              "Unemployment":unemployment})

df2 = census_summary.reset_index(level=['State'])
df2.head(20)
```

```python
#Merge datsets

df_merge = pd.merge(df1, df2, on='State')

df_merge
```

```python
# function to remove non-ASCII for Vineyard column
def remove_non_ascii(Vineyard):
    return ''.join(i for i in Vineyard if ord(i)<128)
 
df_merge['Vineyard'] = df_merge['Vineyard'].apply(remove_non_ascii)

```

```python
# function to remove non-ASCII for Winery column
def remove_non_ascii(Winery):
    return ''.join(i for i in Winery if ord(i)<128)
 
df_merge['Winery'] = df_merge['Winery'].apply(remove_non_ascii)
```

```python
new_df = df_merge[["State", "Vineyard", "Winery", "Cost (Per Bottle)", "Category", "TotalPop"]]
new_df
```

```python
# function to remove non-ASCII for Vineyard column
def remove_non_ascii(Vineyard):
    return ''.join(i for i in Vineyard if ord(i)<128)
 
new_df['Vineyard'] = new_df['Vineyard'].apply(remove_non_ascii)

```

```python
# function to remove non-ASCII for Winery column
def remove_non_ascii(Winery):
    return ''.join(i for i in Winery if ord(i)<128)
 
new_df['Winery'] = new_df['Winery'].apply(remove_non_ascii)
```

```python
cali = new_df.loc[new_df["State"]=="California"]

avg_cali_bottle = round(cali["Cost (Per Bottle)"].mean(),2)
print(avg_cali_bottle)

cali_wine_types = cali["Category"].value_counts()
#print(cali_wine_types)
len_cali_wine_types = len(cali_wine_types)

cali_vineyards = cali["Vineyard"].value_counts()
#print(cali_vineyards)
len_cali_vineyards = len(cali_vineyards)

cali_wineries = cali["Winery"].value_counts()
#print(cali_wineries)
len_cali_wineries = len(cali_wineries)
```

```python
wash = new_df.loc[new_df["State"]=="Washington"]


avg_wash_bottle = round(wash["Cost (Per Bottle)"].mean(),2)
print(avg_wash_bottle)

wash_wine_types = wash["Category"].value_counts()
#print(cali_wash_types)
len_wash_wine_types =len(wash_wine_types)

wash_vineyards = wash["Vineyard"].value_counts()
#print(wash_vineyards)
len_wash_vineyards = len(wash_vineyards)

wash_wineries = wash["Winery"].value_counts()
#print(wash_wineries)
len_wash_wineries = len(wash_wineries)
```

```python
oregon = new_df.loc[new_df["State"]=="Oregon"]


avg_oregon_bottle = round(oregon["Cost (Per Bottle)"].mean(),2)
print(avg_oregon_bottle)

oregon_wine_types = oregon["Category"].value_counts()
#print(oregon_wine_types)
len_oregon_wine_types = len(oregon_wine_types)

oregon_vineyards = oregon["Vineyard"].value_counts()
#print(oregon_vineyards)
len_oregon_vineyards = len(oregon_vineyards)

oregon_wineries = oregon["Winery"].value_counts()
#print(oregon_wineries)
len_oregon_wineries = len(oregon_wineries)
```

```python
ny = new_df.loc[new_df["State"]=="New York"]


avg_ny_bottle = round(ny["Cost (Per Bottle)"].mean(),2)
print(avg_ny_bottle)

ny_wine_types = ny["Category"].value_counts()
#print(ny_wine_types)
len_ny_wine_types = len(ny_wine_types)

ny_vineyards = ny["Vineyard"].value_counts()
#print(ny_vineyards)
len_ny_vineyards = len(ny_vineyards)

ny_wineries = ny["Winery"].value_counts()
#print(ny_wineries)
len_ny_wineries = len(ny_wineries)
```

```python
US_Wineries = pd.DataFrame({
                            "Total Population" : popperstate,
                            "Vineyards" : [len_cali_vineyards, len_wash_vineyards, len_oregon_vineyards, len_ny_vineyards],
                            "Wineries" : [len_cali_wineries, len_wash_wineries, len_oregon_wineries, len_ny_wineries],
                            "Categories of Wine" : [len_cali_wine_types, len_wash_wine_types, len_oregon_wine_types, len_ny_wine_types],
                            "Avg Cost Per Bottle" : [avg_cali_bottle, avg_wash_bottle, avg_oregon_bottle, avg_ny_bottle]})

US_Wineries["Total Population"] = US_Wineries["Total Population"].map("{:,}".format)
US_Wineries["Vineyards"] = US_Wineries["Vineyards"].map("{:,}".format)
US_Wineries["Wineries"] = US_Wineries["Wineries"].map("{:,}".format)
US_Wineries["Avg Cost Per Bottle"] = US_Wineries["Avg Cost Per Bottle"].map("${}".format)
US_Wineries                                                                                            
```
